```python
import pandas as pd
import numpy as np
from sklearn import datasets
from sklearn.svm import SVC
from sklearn.metrics import classification_report
from sklearn.model_selection import train_test_split
```

```python
iris = datasets.load_iris()
X = iris["data"][:, (2, 3)]  
y = iris["target"]

setosa_or_versicolor = (y == 0) | (y == 1)
X = X[setosa_or_versicolor]
y = y[setosa_or_versicolor]
X=pd.DataFrame(X)
y=pd.DataFrame(y)
y.rename(columns = {0 : 'label'}, inplace = True)
X.rename(columns = {0 : 'x1',1 : 'x2'}, inplace = True)
df=pd.concat([X,y],axis=1)
```

```python
X
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2, random_state=17)
print(X_test.shape)
print(X_train.shape)
print(df.shape)
```

C-Dukungan Vektor Klasifikasi.

Implementasinya didasarkan pada libsvm. Skala waktu kecocokan setidaknya secara kuadrat dengan jumlah sampel dan mungkin tidak praktis di luar puluhan ribu sampel. Untuk kumpulan data besar, pertimbangkan untuk menggunakan LinearSVC atau SGDClassifier sebagai gantinya, mungkin setelah transformator Nystroem.

Dukungan multiclass ditangani sesuai dengan skema satu lawan satu.

Untuk detail tentang formulasi matematis yang tepat dari fungsi kernel yang disediakan dan bagaimana gamma, coef0 and degree saling mempengaruhi, lihat bagian terkait dalam dokumentasi naratif: Fungsi kernel.

Parameters

**Cfloat, default=1.0**

Parameter regularisasi. Kekuatan regularisasi berbanding terbalik dengan C. Harus benar-benar positif. Hukumannya adalah penalti kuadrat l2.

**kernel{‘linear’, ‘poly’, ‘rbf’, ‘sigmoid’, ‘precomputed’} or callable, default=’rbf’**

Menentukan jenis kernel yang akan digunakan dalam algoritma. Jika tidak ada yang diberikan, 'rbf' akan digunakan. Jika callable diberikan, itu digunakan untuk menghitung matriks kernel dari matriks data; matriks itu harus berupa array bentuk (n_samples, n_samples).

**degreeint, default=3**

Derajat fungsi kernel polinomial ('poli'). Diabaikan oleh semua kernel lainnya.

**gamma{‘scale’, ‘auto’} or float, default=’scale’**

Koefisien kernel untuk 'rbf', 'poli' dan 'sigmoid'.
- jika gamma='scale' (default) dilewatkan maka menggunakan 1 / (n_features * X.var()) sebagai nilai gamma,
- jika 'otomatis', menggunakan 1 / n_features.

Diubah di versi 0.22: Nilai default gamma berubah dari 'otomatis' menjadi 'skala'.

**coef0float, default=0.0**

Istilah independen dalam fungsi kernel. Ini hanya signifikan dalam 'poli' dan 'sigmoid'.

**shrinkingbool, default=True**

Apakah akan menggunakan heuristik menyusut. 

**probabilitybool, default=False**

Apakah akan mengaktifkan estimasi probabilitas. Ini harus diaktifkan sebelum memanggil fit, akan memperlambat metode itu karena secara internal menggunakan validasi silang 5 kali lipat, dan predict_proba mungkin tidak konsisten dengan predict. 

**tolfloat, default=1e-3**

Toleransi untuk berhenti kriteria.

**cache_sizefloat, default=200**

Tentukan ukuran cache kernel (dalam MB).

**class_weightdict or ‘balanced’, default=None**

Setel parameter C kelas i ke class_weight[i]*C untuk SVC. Jika tidak diberikan, semua kelas seharusnya memiliki bobot satu. Mode “seimbang” menggunakan nilai y untuk menyesuaikan bobot secara otomatis berbanding terbalik dengan frekuensi kelas dalam data input sebagai n_samples / (n_classes * np.bincount(y)).

**verbosebool, default=False**

Aktifkan keluaran verbose. Perhatikan bahwa pengaturan ini memanfaatkan pengaturan runtime per proses di libsvm yang, jika diaktifkan, mungkin tidak berfungsi dengan baik dalam konteks multithread.

**max_iterint, default=-1**

Batas keras pada iterasi dalam pemecah, atau -1 tanpa batas.

**decision_function_shape{‘ovo’, ‘ovr’}, default=’ovr’**

Apakah akan mengembalikan fungsi keputusan satu-vs-istirahat ('ovr') bentuk (n_samples, n_classes) sebagai semua pengklasifikasi lainnya, atau fungsi keputusan satu-vs-satu ('ovo') asli dari libsvm yang memiliki bentuk (n_samples , n_kelas * (n_kelas - 1) / 2). Namun, perhatikan bahwa secara internal, satu lawan satu ('ovo') selalu digunakan sebagai strategi multi-kelas untuk melatih model; matriks ovr hanya dibangun dari matriks ovo. Parameter diabaikan untuk klasifikasi biner.

**break_tiesbool, default=False**

Jika true, decision_function_shape='ovr', dan jumlah kelas > 2, prediksi akan putus sesuai dengan nilai keyakinan dari fungsi_keputusan; jika tidak, kelas pertama di antara kelas terikat dikembalikan. Harap dicatat bahwa pemutusan ikatan datang dengan biaya komputasi yang relatif tinggi dibandingkan dengan prediksi sederhana.

Regresi-SVM

```python
df = pd.read_csv("https://raw.githubusercontent.com/DianCA26/Dataset/main/CarPrice_Assignment.csv")
```

```python
from sklearn.svm import SVR
from sklearn.svm import LinearSVR
from sklearn.preprocessing import MinMaxScaler, StandardScaler
```

```python
df.head()
```

```python
X = df[['horsepower','wheelbase','enginesize','boreratio','compressionratio','highwaympg','citympg','curbweight','carwidth','carlength']]
y = df['price']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2, random_state=17)
print(X_test.shape)
print(X_train.shape)
print(X.shape)
```

```python
sc=MinMaxScaler()
sc.fit(X_train)
X_train_scaled=sc.transform(X_train)
X_test_scaled=sc.transform(X_test)
```

```python
# pakai regresi dengan polynomial

svm_reg = SVR(kernel="poly", degree=2, C=1000, epsilon=0.1, gamma="scale")
svm_reg.fit(X_train_scaled, y_train)
y_pred_train=svm_reg.predict(X_train_scaled)
y_pred_test=svm_reg.predict(X_test_scaled)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score

print('MAE train = ', mean_absolute_error(y_train, y_pred_train))
print('MAE test = ', mean_absolute_error(y_test, y_pred_test))

print('MSE train = ', mean_squared_error(y_train, y_pred_train))
print('MSE test = ', mean_squared_error(y_test, y_pred_test))

print('RMSE train = ', np.sqrt(mean_squared_error(y_train, y_pred_train)))
print('RMSE test = ', np.sqrt(mean_squared_error(y_test, y_pred_test)))

print('r2score train = ', r2_score(y_train, y_pred_train))
print('r2score test = ', r2_score(y_test, y_pred_test))
```

```python
# pakai regresi dengan polynomial yang C = 0.01

svm_reg2 = SVR(kernel="poly", degree=2, C=0.01, epsilon=0.1, gamma="scale")
svm_reg2.fit(X_train_scaled, y_train)
y_pred_train=svm_reg2.predict(X_train_scaled)
y_pred_test=svm_reg2.predict(X_test_scaled)
```

```python
print('MAE train = ', mean_absolute_error(y_train, y_pred_train))
print('MAE test = ', mean_absolute_error(y_test, y_pred_test))

print('MSE train = ', mean_squared_error(y_train, y_pred_train))
print('MSE test = ', mean_squared_error(y_test, y_pred_test))

print('RMSE train = ', np.sqrt(mean_squared_error(y_train, y_pred_train)))
print('RMSE test = ', np.sqrt(mean_squared_error(y_test, y_pred_test)))

print('r2score train = ', r2_score(y_train, y_pred_train))
print('r2score test = ', r2_score(y_test, y_pred_test))
```

```python
# pakai regresi dengan lineaar

mod_SVR=SVR(kernel='linear', epsilon=0.5)
mod_SVR.fit(X_train_scaled,y_train)

y_pred_train=mod_SVR.predict(X_train_scaled)
y_pred_test=mod_SVR.predict(X_test_scaled)
```

```python
print('MAE train = ', mean_absolute_error(y_train, y_pred_train))
print('MAE test = ', mean_absolute_error(y_test, y_pred_test))

print('MSE train = ', mean_squared_error(y_train, y_pred_train))
print('MSE test = ', mean_squared_error(y_test, y_pred_test))

print('RMSE train = ', np.sqrt(mean_squared_error(y_train, y_pred_train)))
print('RMSE test = ', np.sqrt(mean_squared_error(y_test, y_pred_test)))

print('r2score train = ', r2_score(y_train, y_pred_train))
print('r2score test = ', r2_score(y_test, y_pred_test))
```
